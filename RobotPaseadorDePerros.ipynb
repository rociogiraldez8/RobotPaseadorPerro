```python
# --- TRADUCCIÓN DE PSEUDOCÓDIGO A PYTHON REAL ---


import random # Para simular eventos como charcos o necesidades


def simular_robot_paseador():
    # 1. Definición de variables (Subtarea A)
    bateria_actual = 100
    tiempo_transcurrido = 0
    tiempo_max_paseo = 10 # Paseo de 10 minutos para la prueba
    umbral_bateria_critica = 15
    umbral_tiempo_regreso = 3 # Empezar a volver 3 minutos antes
    bolsitas_recogidas = 0


    # Verificación inicial
    if bateria_actual < umbral_bateria_critica:
        print("Robot: Batería demasiado baja para salir.")
        return


    print("Robot: Perro enganchado. ¡Iniciando paseo!")
    print("Robot: Planificando ruta hacia el parque...")


    # 2. Navegación Activa (Subtarea B - Bucle principal)
    # Mientras el tiempo no se agote y la batería sea suficiente:
    while tiempo_transcurrido < tiempo_max_paseo and bateria_actual > umbral_bateria_critica:


        tiempo_transcurrido += 1
        bateria_actual -= 2 # El robot gasta energía
        print(f"\n[Minuto {tiempo_transcurrido}] Paseando por el parque...")


        # 3. Gestión de Incidentes (Subtarea C)


        # Simular detección de charco (20% de probabilidad)
        if random.random() < 0.2:
            print("Robot: ¡Charco detectado! Desviando ruta para no mojar las patas.")


        # Simular necesidad del perro (15% de probabilidad)
        if random.random() < 0.15:
            print("Robot: El perro ha hecho una necesidad. Iniciando recolección...")
            bolsitas_recogidas += 1
            print(f"Robot: Bolsita número {bolsitas_recogidas} almacenada.")


        # 4. Decisión de regreso (Subtarea D)
        tiempo_restante = tiempo_max_paseo - tiempo_transcurrido


        if tiempo_restante <= umbral_tiempo_regreso:
            print(">>> Robot: Tiempo casi agotado. Iniciando retorno a casa.")
            break # Salimos del bucle para ir a la sección de llegada


    # 5. Cierre del Paseo
    print("\n---------------------------------------")
    print("Robot: ¡Hemos llegado a casa!")
    print(f"RESUMEN: {bolsitas_recogidas} necesidades recogidas.")
    print(f"BATERÍA FINAL: {bateria_actual}%")
    print("Robot: Desenganchando correa. Notificación enviada al dueño.")


# Ejecutar la simulación
simular_robot_paseador()
```

```python
import math
import time
import random

class RobotPaseadorPro:
    def __init__(self, nombre, bateria, num_bolsas=5):
        # --- Atributos de Configuración ---
        self.nombre = nombre
        self.bateria = bateria
        self.capacidad_max_bolsas = 10
        self.num_bolsas = num_bolsas
        self.deposito_desechos = 0

        # --- Atributos de Estado y Ubicación ---
        self.pos_actual = (random.randint(1, 10), random.randint(1, 10))
        self.pos_central = (0, 0)
        self.ubicacion_tipo = "Vereda"
        self.clima = "Despejado"
        self.db_eventos = []

    def registrar_en_db(self, evento, detalle):
        self.db_eventos.append({"evento": evento, "detalle": detalle, "hora": time.ctime()})

    def iniciar_sistema(self):
        print(f"--- 🤖 Iniciando {self.nombre} ---")

        # 1. Verificar batería
        if self.bateria < 50:
            print(f"Batería baja ({self.bateria}%). Cargando...")
            self.bateria = 100

        # 2. Configurar rutas y verificar elementos
        print(f"Configurando rutas. Posición actual: {self.pos_actual}")
        if self.num_bolsas == 0:
            print("❌ ERROR: Sin bolsas. No se puede iniciar el paseo.")
            return False

        print("✅ Sistema listo. Iniciando paseo por la vereda.")
        return True

    def caminar(self, obstaculo=None):
        # Regla de oro: Siempre por la vereda
        if self.ubicacion_tipo != "Vereda":
            self.ubicacion_tipo = "Vereda"
            print("🔄 Corrigiendo: El robot ha regresado a la vereda.")

        if obstaculo == "Charco":
            print("🌊 Charco detectado. Calculando maniobra de evasión...")
            self.registrar_en_db("Obstáculo", "Charco evitado")
        elif obstaculo == "Otro perro":
            print("🐕 Contacto con otro perro. Manteniendo curso planificado.")
        else:
            print("🚶 Caminando tranquilamente...")

    def gestionar_desechos(self, certeza_ia):
        print("⚠️ Posible desecho detectado por sensores visuales.")

        # Lógica de Probabilidad (Inferencia)
        if certeza_ia < 0.8:
            print(f"🤔 Certeza del {certeza_ia*100}%: Parece una piedra. Ignorando.")
            return

        # Lógica de Recursos (Bolsas)
        if self.num_bolsas > 0:
            print("💩 Ejecutando protocolo de recogida:")
            print("   [Sacando dispensador -> Tomando bolsa -> Inclinándose -> Cerrando bolsa]")
            self.num_bolsas -= 1
            self.deposito_desechos += 1
            self.registrar_en_db("Limpieza", "Desecho recogido")
        else:
            print("❌ ALERTA: No quedan bolsas en el inventario.")

    def verificar_clima(self, estado_clima):
        self.clima = estado_clima
        if self.clima == "Lluvia":
            print("🌧️ ¡Alerta de lluvia! Resguardando al perro en lugar techado...")
            print("... Esperando a que el sensor de humedad baje ...")
            time.sleep(1)
            self.clima = "Despejado"
            print("🌤️ Clima óptimo. Reanudando servicio.")

    # --- NUEVAS FUNCIONES DE OPTIMIZACIÓN Y CIERRE ---

    def calcular_ruta_optima(self):
        """Usa la Distancia Euclídea para calcular el regreso"""
        x1, y1 = self.pos_actual
        x2, y2 = self.pos_central
        distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        print(f"📍 Calculando ruta más corta a Central. Distancia: {distancia:.2f} metros.")
        return distancia

    def finalizar_mision(self):
        print("\n--- INICIANDO PROTOCOLO DE CIERRE ---")

        # 1. Volver a la central
        dist = self.calcular_ruta_optima()
        self.pos_actual = self.pos_central
        print(f"🏠 El robot ha recorrido {dist:.2f}m y ha llegado a la Central.")

        # 2. Descartar desechos
        if self.deposito_desechos > 0:
            print(f"🗑️ Vaciando depósito: {self.deposito_desechos} bolsas descartadas.")
            self.deposito_desechos = 0

        # 3. Recargar batería
        self.bateria = 100
        print("⚡ Batería recargada al 100%.")

        # 4. Reto: Calcular reposición de bolsas
        bolsas_a_comprar = self.capacidad_max_bolsas - self.num_bolsas
        print(f"📦 Inventario: Tienes {self.num_bolsas} bolsas. Debes reponer {bolsas_a_comprar} para el próximo turno.")

# --- SIMULACIÓN DEL PASEO ---

robot = RobotPaseadorPro("Pythoniso-Bot", bateria=40, num_bolsas=4)

if robot.iniciar_sistema():
    robot.caminar(obstaculo="Charco")
    robot.gestionar_desechos(certeza_ia=0.95) # Es un desecho real
    robot.verificar_clima("Lluvia")
    robot.caminar(obstaculo="Otro perro")
    robot.finalizar_mision()
```

```python
import random
import time
import math
from datetime import datetime

class RobotMadridPro:
    def __init__(self, nombre_robot, bateria):
        self.nombre_robot = nombre_robot
        self.bateria = bateria
        self.num_bolsas = 10
        self.nivel_agua = 100  # Porcentaje del tanque
        self.deposito_desechos = 0
        self.archivo_historial = "historial_paseos.txt"

        # Memoria de perros conocidos
        self.base_perros = {
            "Tobi": {"raza": "Golden", "edad": 3, "paseos": 5},
            "Luna": {"raza": "Galgo", "edad": 5, "paseos": 12}
        }

        # Variables de estado del paseo actual
        self.nombre_perro = ""
        self.perro_actual = {}
        self.destino = ""
        self.log_paseo = []
        self.clima = random.choice(["Soleado", "Nublado", "Caluroso"])

    def registrar_evento(self, mensaje):
        """Registra hitos con hora exacta"""
        hora = datetime.now().strftime("%H:%M:%S")
        entrada = f"[{hora}] {mensaje}"
        self.log_paseo.append(entrada)

    def identificar_y_planificar(self):
        """Entrevista inicial y ficha técnica"""
        print(f"\n🤖 [SISTEMA]: {self.nombre_robot} activo en La Latina.")
        nombre = input("👤 [DUEÑO]: ¿Nombre del perro? -> ").strip().title()
        self.nombre_perro = nombre

        # Registro de perro nuevo o conocido
        if nombre not in self.base_perros:
            print(f"🤖 [ROBOT]: {nombre} es nuevo. Creando ficha...")
            raza = input("👤 [DUEÑO]: ¿Raza? -> ")
            edad = int(input("👤 [DUEÑO]: ¿Edad? -> "))
            self.base_perros[nombre] = {"raza": raza, "edad": edad, "paseos": 0}

        self.perro_actual = self.base_perros[nombre]

        # Mostrar Ficha
        print(f"\n📋 FICHA: {nombre} | {self.perro_actual['raza']} | {self.perro_actual['edad']} años.")

        # Planificación
        self.destino = input("👤 [DUEÑO]: ¿Destino del paseo? -> ").strip().title()
        decision = input("👤 [DUEÑO]: ¿Salir ahora o programar? (ahora/luego) -> ").lower()

        if "lue" in decision:
            print("⏰ [AGENDA]: Paseo programado. Esperando turno...")
            time.sleep(2) # Simulación de espera

        return True

    def gestionar_agua(self, accion):
        """Dispensador inteligente: Hidratación y Limpieza de Pis"""
        gasto = 15 if accion == "pis" else 25

        if self.nivel_agua >= gasto:
            self.nivel_agua -= gasto
            if accion == "pis":
                self.registrar_evento(f"💧 [LIMPIEZA]: Orín limpiado. Tanque al {self.nivel_agua}%")
            else:
                self.registrar_evento(f"🐕 [SALUD]: Perro hidratado. Tanque al {self.nivel_agua}%")
        else:
            self.registrar_evento("⚠️ [ALERTA]: Sin agua suficiente para la acción.")

    def ejecutar_paseo(self):
        """Recorrido paso a paso por Madrid"""
        print(f"\n--- 🚶 INICIANDO PASEO: {self.nombre_perro} ➔ {self.destino} ---")
        self.registrar_evento(f"INICIO: Clima {self.clima}. Salida desde La Cebada.")

        ruta = ["Calle Toledo", "Plaza Mayor", "Calle Alcalá", "Llegada al Destino"]

        for etapa in ruta:
            time.sleep(1) # Simulación de movimiento
            self.registrar_evento(f"Transitand por {etapa} (Seguridad DGT).")

            # Probabilidades de imprevistos
            azar = random.random()
            if azar < 0.15: self.gestionar_agua("pis")
            if 0.15 <= azar < 0.30:
                self.registrar_evento("💩 Desechos sólidos recogidos.")
                self.deposito_desechos += 1
                self.num_bolsas -= 1
            if self.clima == "Caluroso" and azar > 0.8:
                self.gestionar_agua("hidratar")

    def finalizar_y_guardar(self):
        """Reporte final y guardado en archivo .txt"""
        self.perro_actual["paseos"] += 1

        print("\n" + "="*45)
        print(f"📊 REPORTE DE PASEO PARA EL DUEÑO")
        for hito in self.log_paseo:
            print(hito)
        print("="*45)

        # Guardado Persistente
        try:
            with open(self.archivo_historial, "a", encoding="utf-8") as f:
                f.write(f"\nREGISTRO: {datetime.now().strftime('%d/%m/%Y')}\n")
                f.write(f"Perro: {self.nombre_perro} | Destino: {self.destino}\n")
                f.writelines(line + "\n" for line in self.log_paseo)
                f.write("-" * 20 + "\n")
            print(f"💾 Historial actualizado en '{self.archivo_historial}'.")
        except Exception as e:
            print(f"❌ Error al guardar: {e}")

        # Feedback Final
        input(f"\n🤖 [ROBOT]: Entrego a {self.nombre_perro}. ¿Qué tal lo ve? -> ")
        print("🤖 [ROBOT]: Gracias. Volviendo a central para recarga total.")

# --- INICIO DEL PROGRAMA ---
bot = RobotMadridPro("Chulapo-Pro-V4", bateria=100)

if bot.identificar_y_planificar():
    bot.ejecutar_paseo()
    bot.finalizar_y_guardar()
```
